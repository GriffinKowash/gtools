```python
import numpy as np
import matplotlib.pyplot as plt

from gtools import gtools
```

```python
t = [0, 1, 1.5, 1.7, 2.3, 2.5, 3, 5, 7.7, 8, 9.9, 10, 11, 13, 14]
x = np.sin(t)
```

```python
t_res1, x_res1 = gtools.resample(t, x, 0.1)
t_res2, x_res2 = gtools.resample(t, x, 2)
```

```python
plt.plot(t, x)
plt.scatter(t_res1, x_res1, color='C1')
plt.scatter(t_res2, x_res2, color='C2')
```
