```python
import numpy as np
import matplotlib.pyplot as plt

from gtools import gtools
```

```python
### Test of FFT functionality ###

# Test of 1D data array
t = np.linspace(0, 20*2*np.pi, 100000)
x = np.sum([np.sin(w*t) for w in range(1, 20)], axis=0)

f, xfft = gtools.rfft(t, x)

plt.semilogx(f, np.abs(xfft))
```

```python
# Test of 2D data array
x2 = np.sum([np.sin(ww*t) for ww in [2, 3, 4]], axis=0)
X = np.stack([x, x2])

ff, Xfft = gtools.rfft(t, X)

fig, (ax1, ax2) = plt.subplots(2)
ax1.semilogx(ff, np.abs(Xfft[0]))
ax2.semilogx(ff, np.abs(Xfft[1]))
```

```python
### Test of exceptions and warnings ###

# Test: >1d time array exception
t2d = np.ndarray((2,100000))
gtools.rfft(t2d, x)
```

```python
# Test: >1d time array exception
t_wrongshape = np.ndarray(2000)
gtools.rfft(t_wrongshape, x)
```
